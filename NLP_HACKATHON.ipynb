```python
############## text cleaning and preprocessing ###################################
import pandas as pd
import re
import spacy
from datasets import load_dataset
from sklearn.utils import resample
from gensim.models import Word2Vec

# Load the ADE classification dataset
dm = load_dataset("ade-benchmark-corpus/ade_corpus_v2", "Ade_corpus_v2_classification")

dm_df = pd.DataFrame(dm['train'])

# Step 1: Text Cleaning
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))  # Remove special characters
    text = text.lower().strip()  # Convert to lowercase
    return text

dm_df['text'] = dm_df['text'].apply(clean_text)

# Step 2: Tokenization (Using SpaCy general English model)
nlp = spacy.load("en_core_web_sm")  # General English model
dm_df['tokens'] = dm_df['text'].apply(lambda x: [token.text for token in nlp(x)])

# Step 3: Normalization (Handling abbreviations)
medical_abbr = {"htn": "hypertension", "dm": "diabetes mellitus", "mi": "myocardial infarction"}  # Example
dm_df['tokens'] = dm_df['tokens'].apply(lambda tokens: [medical_abbr.get(token, token) for token in tokens])

# Step 4: Annotation Handling
dm_df['relations'] = dm_df.apply(lambda row: {
    'drug': row.get('drug', ''),
    'effect': row.get('effect', ''),
    'relation': row.get('label', '')
}, axis=1)

# Step 5: Class Balancing (Using Oversampling)
majority_class = dm_df[dm_df['label'] == 0]  # Non-ADE samples
minority_class = dm_df[dm_df['label'] == 1]  # ADE samples

minority_upsampled = resample(minority_class,
                              replace=True,
                              n_samples=len(majority_class),
                              random_state=42)

balanced_df = pd.concat([majority_class, minority_upsampled])

# Step 6: Word Embedding (Word2Vec Example)
model = Word2Vec(sentences=balanced_df['tokens'], vector_size=100, window=5, min_count=1, workers=4)

balanced_df['embeddings'] = balanced_df['tokens'].apply(lambda tokens: [model.wv[token] for token in tokens if token in model.wv])

# Display the processed data
print(balanced_df.head())
```

```python
##### CLINICAL MODEL TRAINING#########
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from sklearn.metrics import accuracy_score, classification_report
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
import pandas as pd
from datasets import load_dataset
import re

# Load the ADE classification dataset
dm = load_dataset("ade-benchmark-corpus/ade_corpus_v2", "Ade_corpus_v2_classification")
dm_df = pd.DataFrame(dm['train'])

# Ensure correct column names
print(dm_df.columns)  # Debugging step
if 'text' not in dm_df.columns or 'label' not in dm_df.columns:
    raise KeyError("Columns 'text' and 'label' not found in dataset.")

# Step 1: Text Cleaning
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))  # Remove special characters
    text = text.lower().strip()  # Convert to lowercase
    return text

dm_df['text'] = dm_df['text'].apply(clean_text)

# Load ClinicalBERT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
model = AutoModelForSequenceClassification.from_pretrained("emilyalsentzer/Bio_ClinicalBERT", num_labels=2)

dev = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(dev)

# Tokenization
def tokenize_data(texts, labels, tokenizer, max_length=512):
    encoding = tokenizer(texts.tolist(), padding=True, truncation=True, max_length=max_length, return_tensors="pt")
    return TensorDataset(encoding['input_ids'], encoding['attention_mask'], torch.tensor(labels.values))

X_train, X_test, y_train, y_test = train_test_split(dm_df['text'], dm_df['label'], test_size=0.2, random_state=42)

train_dataset = tokenize_data(X_train, y_train, tokenizer)
test_dataset = tokenize_data(X_test, y_test, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)

# Training the ClinicalBERT model
def train_model(model, train_loader, device):
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
    loss_fn = torch.nn.CrossEntropyLoss()

    model.train()
    for epoch in range(3):
        total_loss = 0
        for batch in train_loader:
            input_ids, attention_mask, labels = [x.to(device) for x in batch]

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs.logits, labels)

            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch + 1}, Loss: {total_loss / len(train_loader):.4f}")

# Evaluation
def evaluate_model(model, test_loader, device):
    model.eval()
    preds, true_labels = [], []

    with torch.no_grad():
        for batch in test_loader:
            input_ids, attention_mask, labels = [x.to(device) for x in batch]

            outputs = model(input_ids, attention_mask=attention_mask)
            preds.extend(torch.argmax(outputs.logits, axis=1).cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    print("ClinicalBERT Accuracy:", accuracy_score(true_labels, preds))
    print(classification_report(true_labels, preds))

# Sample Prediction Function
def predict_sample(text):
    model.eval()
    encoding = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512).to(dev)
    with torch.no_grad():
        output = model(**encoding)
        prediction = torch.argmax(output.logits, axis=1).item()
    print(f"Input: {text}\nPrediction: {['No ADE', 'ADE'][prediction]}")

# Train and Evaluate
train_model(model, train_loader, dev)
evaluate_model(model, test_loader, dev)

# Sample Inputs
sample_texts = [
    "Patient experienced severe rash after taking the medication.",
    "No side effects reported during the medication trial.",
    "Mild dizziness observed after medication usage."
]
for text in sample_texts:
    predict_sample(text)
```
